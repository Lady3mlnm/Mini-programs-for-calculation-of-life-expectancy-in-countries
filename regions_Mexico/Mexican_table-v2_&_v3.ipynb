```python
df = pd.read_excel('data/Mortalidad_09.xlsx', sheet_name='Tabulado', skiprows=6, skipfooter=11,
                   header=None, index_col=0) # skiprows=4, header=None, index_col=0)
df.index.name = ''

df.dropna(how='all', inplace=True)

df.rename(index = {'Estados Unidos Mexicanos': '=MÉXICO=',
                   'Ciudad de México': 'Mexico City',
                   'Coahuila de Zaragoza': 'Coahuila',
                   'Michoacán de Ocampo': 'Michoacán',
                   'México': 'State of México',
                   'Veracruz de Ignacio de la Llave': 'Veracruz'}, inplace=True)

df.head(3)
```

```python
ls_cols = []
for year in list(range(2010, 2025)):
    ls_cols.extend([f'{year}_t', f'{year}_m', f'{year}_f'])

df.columns = ls_cols
df.head(3)
```

```python
def sort_df(df, by=['2023_t', '2023_m', '2023_f'], ascending=False):
    df.sort_values(by=by, ascending=ascending, inplace=True)
    indexes = df.index.to_list()
    ls_change_order = ['=MÉXICO=']
    indexes = list(el for el in indexes if el not in ls_change_order)
    indexes = ls_change_order + indexes
    return df.reindex(indexes)

df = sort_df(df)
df.head(3)
```

```python
def extend_df(df, prec=1):
    df.insert(loc=1, column='2014→2019', value=(df['2019_t']-df['2014_t']).round(prec))
    df.insert(loc=3, column='2019→2020', value=(df['2020_t']-df['2019_t']).round(prec))
    df.insert(loc=5, column='2020→2021', value=(df['2021_t']-df['2020_t']).round(prec))
    df.insert(loc=7, column='2021→2022', value=(df['2022_t']-df['2021_t']).round(prec))
    df.insert(loc=9, column='2022→2023', value=(df['2023_t']-df['2022_t']).round(prec))
    df.insert(loc=11, column='2019→2023', value=(df['2023_t']-df['2019_t']).round(prec))
    df.insert(loc=12, column='2014→2023', value=(df['2023_t']-df['2014_t']).round(prec))
    df['2023_fΔm']  = (df['2023_f']-df['2023_m']).round(prec)
    # df['2019→2023'] = (df['2023_t']-df['2019_t']).round(prec)
    
    return df
    
df = extend_df(df)
df
```

```python
# create code for placing info in Wikipedia
def create_table_regions_v2(df, file_header, lang='ru'):

    def if_value(x, prec=1):
        return '—' if math.isnan(x) else \
               f"{x:0.{prec}f}"  if x>=0 else \
               f"−{-x:0.{prec}f}"                #"{x:0.{prec}f}".format(x, prec)
    
    def chval(x, prec=1):  # change_value
        return '"| —' if math.isnan(x) else \
               f' color:darkgreen;"| {x:0.{prec}f}' if x>0 else \
               f' color:crimson;"| −{-x:0.{prec}f}' if x<0 else \
               f' color:darkgray;"| {x:0.{prec}f}'
    
    def chval_bold(x, prec=1):  # change_value
        return '"| —' if math.isnan(x) else \
               f' color:darkgreen;"| \'\'\'{x:0.{prec}f}\'\'\'' if x>0 else \
               f' color:crimson;"| \'\'\'−{-x:0.{prec}f}\'\'\'' if x<0 else \
               f' color:darkgray;"| \'\'\'{x:0.{prec}f}\'\'\''

    with open('design/' + file_header, mode='r', encoding="utf-8") as fh:
        table_header = fh.read()

    st = ''
    for i in range(len(df)):
        ser = df.iloc[i]    
        if ser.name == '=MÉXICO=':
             st += '\n' + '|-class=static-row-header\n' + \
                  f'|style="text-align:left;"|\'\'\'{dd_replacement_regions[ser.name][lang][0]}\'\'\' ' + \
                  f'|| style="text-align:center;background:#e0ffd8;"| \'\'\'{if_value(ser["2023_t"])}\'\'\' ' + \
                  f'|| style="text-align:center;background:#eaf3ff;"| \'\'\'{if_value(ser["2023_m"])}\'\'\' ' + \
                  f'|| style="text-align:center;background:#fee7f6;"| \'\'\'{if_value(ser["2023_f"])}\'\'\' ' + \
                  f'|| style="text-align:center;background:#fff8dc;"| \'\'\'{if_value(ser["2023_fΔm"])}\'\'\' ' + \
                  f'|| style="text-align:center;border-left-width:2px;padding-left:1em;"| \'\'\'{if_value(ser["2014_t"])}\'\'\' ' + \
                  f'|| style="background:#fffae0;{chval_bold(ser["2014→2019"])} ' + \
                  f'|| style="text-align:center;"| \'\'\'{if_value(ser["2019_t"])}\'\'\' ' + \
                  f'|| style="background:#fffae0;{chval_bold(ser["2019→2020"])} ' + \
                  f'|| style="text-align:center;"| \'\'\'{if_value(ser["2020_t"])}\'\'\' ' + \
                  f'|| style="background:#fffae0;{chval_bold(ser["2020→2021"])} ' + \
                  f'|| style="text-align:center;"| \'\'\'{if_value(ser["2021_t"])}\'\'\' ' + \
                  f'|| style="background:#fffae0;{chval_bold(ser["2021→2022"])} ' + \
                  f'|| style="text-align:center;"| \'\'\'{if_value(ser["2022_t"])}\'\'\' ' + \
                  f'|| style="background:#fffae0;{chval_bold(ser["2022→2023"])} ' + \
                  f'|| style="text-align:center;background:#e0ffd8;"| \'\'\'{if_value(ser["2023_t"])}\'\'\' ' + \
                  f'||style="background:#fffae0; border-left-width:2px;{chval_bold(ser["2019→2023"])}' + \
                  (f' ||style="background:#fffae0;{chval_bold(ser["2014→2023"])}' if lang=='ru' else '')

        else:
            name_link = dd_replacement_regions[ser.name][lang][1]
            name_visible = dd_replacement_regions[ser.name][lang][0]
            name_inserted = name_link if name_link == name_visible else f"{name_link}|{name_visible}"
            st += '\n' + '|-\n' + \
                  f'|style="text-align:left;"| [[{name_inserted}]] ' + \
                  f'|| style="text-align:center;background:#e0ffd8;"| \'\'\'{if_value(ser["2023_t"])}\'\'\' ' + \
                  f'|| style="text-align:center;background:#eaf3ff;"| {if_value(ser["2023_m"])} ' + \
                  f'|| style="text-align:center;background:#fee7f6;"| {if_value(ser["2023_f"])} ' + \
                  f'|| style="text-align:center;background:#fff8dc;"| {if_value(ser["2023_fΔm"])} ' + \
                  f'|| style="text-align:center;border-left-width:2px;padding-left:1em;"| {if_value(ser["2014_t"])} ' + \
                  f'|| style="background:#fffae0;{chval(ser["2014→2019"])} ' + \
                  f'|| style="text-align:center;"| {if_value(ser["2019_t"])} ' + \
                  f'|| style="background:#fffae0;{chval(ser["2019→2020"])} ' + \
                  f'|| style="text-align:center;"| {if_value(ser["2020_t"])} ' + \
                  f'|| style="background:#fffae0;{chval(ser["2020→2021"])} ' + \
                  f'|| style="text-align:center;"| {if_value(ser["2021_t"])} ' + \
                  f'|| style="background:#fffae0;{chval(ser["2021→2022"])} ' + \
                  f'|| style="text-align:center;"| {if_value(ser["2022_t"])} ' + \
                  f'|| style="background:#fffae0;{chval(ser["2022→2023"])} ' + \
                  f'|| style="text-align:center;background:#e0ffd8;"| \'\'\'{if_value(ser["2023_t"])}\'\'\' ' + \
                  f'||style="background:#fffae0; border-left-width:2px;{chval(ser["2019→2023"])}' + \
                  (f' ||style="background:#fffae0;{chval(ser["2014→2023"])}' if lang=='ru' else '')

    if lang == 'ru':
        st = re.sub('(?<=\\d)\\.(?=\\d)', ',', st)  # replace . to comma, if this . is between two digits
        st = st.replace('padding-right:1,5ex;', 'padding-right:1.5ex;')

    st = table_header + st + '\n|}'
    
    # gray color for missing values
    st = st.replace(';"| —', ';color:silver;"| —')

    return st


table_code = create_table_regions_v2(df, file_header='Mexican_states_header_ru -2023 -v2.txt', lang='ru')

# write the code to file
with open('output/Table code for Mexican states -v2 -ru.txt', 'w', encoding="utf-8") as fh:
    fh.write(table_code)
```
