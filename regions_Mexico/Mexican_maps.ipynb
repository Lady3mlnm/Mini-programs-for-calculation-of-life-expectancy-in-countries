```python
def sort_df(df, by=['2023_t', '2022_t'], ascending=False):
    df.sort_values(by=by, ascending=ascending, inplace=True)
    indexes = df.index.to_list()
    ls_change_order = ['=MÉXICO=']
    indexes = list(el for el in indexes if el not in ls_change_order)
    indexes = ls_change_order + indexes
    return df.reindex(indexes)

df = sort_df(df)
df.head(3)
```

```python
SELECTED_COL = 2023
df_grouped = mal.bin_values_in_dataframe(df_map, SELECTED_COL, step = 0.5, prec=1)

# df_grouped['group_label'] = df_grouped['group_label'].map(lambda st: '70.0–74.99' if st < '75.0–75.49' else st)

df_grouped
```

```python
def extract_indexes(subdf, dd_legend = dd_legend):
    group_label = subdf['group_label'].iloc[0]
    
    regions = subdf.index.to_list()
    color = (dd_legend[group_label])
    
    ls_grouping.append(RegionGroup(group_label=group_label, regions=regions, color=color))

    return pd.Series([color, regions], index=['color', 'regions'])


ls_grouping = []
df_grouped = df_grouped.groupby(['group_label'])[['group_label']].apply(extract_indexes).loc[::-1]

df_grouped
```

```python
def create_json(ls_grouping, title=''):
    false, true = False, True
    
    jo = {
        "groups": { },
        "title": title,
        "hidden": [],
        "background": "#ffffff",
        "borders": "#000",
        "legendFont": "Century Gothic",
        "legendFontColor": "#000",
        "legendBorderColor": "#00000000",
        "legendBgColor": "#00000000",
        "legendWidth": 121.14347839355469,
        "legendBoxShape": "square",
        "areBordersShown": true,
        "defaultColor": "#d1dbdd",
        "labelsColor": "#6a0707",
        "labelsFont": "Arial",
        "strokeWidth": "medium",
        "areLabelsShown": true,
        "uncoloredScriptColor": "#ffff33",
        "v6": true,
        "levelsVisibility": [
            "show",
            "show",
            "transparent"
        ],
        "legendPosition": "custom",
        "legendX": 747.7223919785803,
        "legendY": 30.466462110793174,
        "legendSize": "large",
        "legendStatus": "show",
        "scalingPatterns": true,
        "legendRowsSameColor": true,
        "legendColumnCount": 1
    }
    
    for group_label, color, regions in ls_grouping[::-1]:
        jo["groups"][f"#{color}"] = {"label": group_label,
                                     "paths": [region.replace(' ', '_').replace('Labrador', 'Labrador_') for region in regions]}       

    return jo


jo = create_json(ls_grouping, title=SELECTED_COL)

pretty_jo = json.dumps(jo, indent=2, ensure_ascii=False)

with open(f"output/map_JSON -{SELECTED_COL}.txt", 'w', encoding="utf-8") as fh:
    fh.write(pretty_jo)

# print(pretty_jo)
```
