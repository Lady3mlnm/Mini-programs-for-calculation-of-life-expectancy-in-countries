```python
def create_svg_gend(df, districts, colors, box_width=147, box_height=236, svg_width=357, file_name='le_diagram_WHO -test.svg', lang='ru'):
    outs = {k: '' for k in districts.keys()}

    out_defs = out_graph = ''

    for i in range(len(df)):
        # extract single values for country
        le_a, le_m, le_f, _, _, population, district, le_um, le_uf, le_rm, le_rf, x_text, y_text, text_anchor = stat.iloc[i]
        
        name = df.iloc[i].name

        # transtlate name of country to Russian if this is required
        name = name if lang=='ru' else dd_to_eng[name]

        x = int(le_m * 10)
        y = int(le_f * -10)
        radius = round(population ** (1/2) / 300, 2)
        color = colors[district]

        if pd.notna(text_anchor):  # if there is label for country
            out_label_defs = f'\n      <text x="{x_text}" y="{y_text}" dy="0.7ex" text-anchor="{text_anchor}" stroke="none">{name}</text>'
            out_label_active = "";
        else:
            x_text = int(x - radius * 0.8);
            y_text = int(y - radius * 0.8);
            out_label_defs = "";
            out_label_active = f'\n      <text x="{x_text}" y="{y_text}" dy="0.7ex" class="hidden_text" stroke="{color}">{name}</text>'

        population = f"{population:,}".replace(',', ' ')

        out_defs += (
            f'\n    <g id="bubble_{i}" fill="{color}">{out_label_defs}'
            f'\n      <circle cx="{x}" cy="{y}" r="1"/>'
            f'\n      <circle cx="{x}" cy="{y}" r="{radius}" stroke="{color}" fill="url(#grad_{district})"/>'
            f'\n    </g>')

        out_graph += \
            f'\n    <g class="active">' + \
            f'\n      <use xlink:href="#bubble_{i}"/>{out_label_active}' + \
           (f'\n      <title>{dd_decryption_ru.get(name, name)} (население: {population})' +
            f'\n\nобщая: {le_a} лет' +
            f'\n\nмужчины: {le_m}  (в городах: {le_um}, в сельской местности: {le_rm})' +
            f'\n\nженщины: {le_f}  (в городах: {le_uf}, в сельской местности: {le_rf})</title>' if lang=='ru' else
            f'\n      <title>{dd_decryption_en.get(name, name)} (population: {population})' 
            f'\n\nall: {le_a} years' +
            f'\n\nmale: {le_m}  (in urban: {le_um}, in rural: {le_rm})' +
            f'\n\nfemale: {le_f}  (in urban: {le_uf}, in rural: {le_rf})</title>') + \
            f'\n    </g>'

        outs[district] += f'\n        <use xlink:href="#bubble_{i}"/>'

    ##########
    out_grad = out_legend = ''
    y = -611
    x_legend = 510

    for i, abbr in enumerate(districts.keys(), start=1):
        color = colors[abbr]
        district = districts[abbr]
        
        # transtlate name of district to English if this is required
        district = district if lang=='ru' else dd_to_eng[district]

        out_grad += (
            f'\n    <radialGradient id="grad_{abbr}" cx="50%" cy="50%" r="50%" fx="25%" fy="25%">'
            f'\n      <stop offset="50%" stop-color="{color}" stop-opacity="0"/>'
            f'\n      <stop offset="70%" stop-color="{color}" stop-opacity="0.05"/>'
            f'\n      <stop offset="99%" stop-color="{color}" stop-opacity="0.25"/>'
            f'\n    </radialGradient>')

        x = y + 5
        out_legend += (
            f'\n      <g class="active">'
            f'\n        <circle cx="{x_legend}" cy="{y}" r="6" fill="url(#grad_{abbr})"/>'
            f'\n        <circle cx="{x_legend}" cy="{y}" r="1" fill="{color}"/>'
            f'\n        <text x="{x_legend+9}" y="{x-1}" fill="{color}">{district}</text>'
            + outs[abbr] +
            f'\n      </g>')

        y += 14
        
        if i == 4:
            y = -611
            x_legend = 672

    #######
    out_ticks = ''

    for tick in range(60, 81, 5):
        x = tick * 10
        out_ticks += f'\n        <text x="{x}" y="-638">{tick}</text>'

    for tick in range(70, 86, 5):
        y = tick * -10 + 6
        out_ticks += f'\n        <text x="490" y="{y}">{tick}</text>'

    ######
    svg_code = """<?xml version="1.0" encoding="utf-8"?>
<svg xmlns="http://www.w3.org/2000/svg" xmlns:xlink="http://www.w3.org/1999/xlink" width="100%" height="100%" viewBox="470 -854 """+ str(svg_width) + """ 297">
  <title>Comparison gender life expectancy WHO-2019</title>
  <desc>Comparison of federal subjects of Russia by life expectancy at birth according to data of Rosstat. Data for life expectancy: https://rosstat.gov.ru/folder/210/document/13207, https://www.fedstat.ru/indicator/31293. Data for number of population: https://rosstat.gov.ru/folder/12781, https://rosstat.gov.ru/storage/mediabank/Popul_Comp2022_site.xls. Code of the program for creation of this chart: https://github.com/Lady3mlnm/Mini-programs-for-calculation-of-life-expectancy-in-countries/tree/main/SVG-diagrams-Russia.</desc>
  <style type="text/css">
    #main         { font-family:Helvetica,Arial,sans-serif; font-size:11px; text-anchor:middle;
                    stroke-opacity:0; fill-opacity:0.5; cursor:default; }
    #main:hover   { stroke-opacity:0; fill-opacity:0.1; }
    .nofade,
    .active:hover { stroke-opacity:1; fill-opacity:1; }
    .hidden_text  { font-size:12px; text-anchor:end; fill:none; pointer-events:none; }
  </style>
  <defs>

    <!-- determine appearance of circles -->""" + \
    out_grad + \
    """

    <!-- determine each country -->""" + \
    out_defs + \
    """

    <!-- vertical grid -->
    <path id="grid_x" d="M 0 -650 V -850"/>
    <g id="grid_x_5">
      <use xlink:href="#grid_x"/>
      <use xlink:href="#grid_x" transform="translate(10,0)"/>
      <use xlink:href="#grid_x" transform="translate(20,0)"/>
      <use xlink:href="#grid_x" transform="translate(30,0)"/>
      <use xlink:href="#grid_x" transform="translate(40,0)"/>
      <use xlink:href="#grid_x" transform="translate(50,0)"/>
    </g>

    <!-- horizontal grid -->
    <path id="grid_y" d="M 500 0 H 800"/>
    <g id="grid_y_5">
      <use xlink:href="#grid_y"/>
      <use xlink:href="#grid_y" transform="translate(0,-10)"/>
      <use xlink:href="#grid_y" transform="translate(0,-20)"/>
      <use xlink:href="#grid_y" transform="translate(0,-30)"/>
      <use xlink:href="#grid_y" transform="translate(0,-40)"/>
      <use xlink:href="#grid_y" transform="translate(0,-50)"/>
    </g>
  </defs>



  <g id="main">
    <g class="nofade">
      <!-- background for the whole diagram -->
      <circle cx="0" cy="0" r="99999" fill="#ffffff"/>

      <!-- draw grid -->
      <g stroke-opacity="0.05" stroke="#000000">
        <use xlink:href="#grid_x_5" transform="translate(500,0)"/>
        <use xlink:href="#grid_x_5" transform="translate(550,0)"/>
        <use xlink:href="#grid_x_5" transform="translate(600,0)"/>
        <use xlink:href="#grid_x_5" transform="translate(650,0)"/>
        <use xlink:href="#grid_x_5" transform="translate(700,0)"/>
        <use xlink:href="#grid_x_5" transform="translate(750,0)"/>
        <use xlink:href="#grid_y_5" transform="translate(0,-650)"/>
        <use xlink:href="#grid_y_5" transform="translate(0,-700)"/>
        <use xlink:href="#grid_y_5" transform="translate(0,-750)"/>
        <use xlink:href="#grid_y_5" transform="translate(0,-800)"/>
      </g>

      <!-- labels for axes -->
      <g fill="#999999">  <!-- color of font -->""" + \
    ("""
        <text x="655" y="-628" style="font-size: 10px">Ожидаемая продолжительность жизни для мужчин</text>
        <text transform="rotate(-90)" x="732" y="479" style="font-size: 10px">Ожидаемая продолжительность жизни для женщин</text>""" if lang=='ru' else 
     """
        <text x="673" y="-628">Male life expectancy at birth in years</text>
        <text transform="rotate(-90)" x="745" y="479">Female life expectancy at birth in years</text>""") + \
        out_ticks + \
    """
      </g>

      <!-- diagonal line -->
      <path d="M 690 -690 L 800 -800" stroke="#00cc00" stroke-width="0.75" stroke-opacity="0.4"/>
    </g>

    <!-- menu -->
    <g text-anchor="start">

      <!-- rectangle around items -->
      <rect class="nofade" x="500" y="-620" width=\"""" + str(box_width) + """\" height=\"""" + str(box_height) + """\" rx="5" ry="5" stroke="#999999" fill="#ffffff"/>

      <!-- single items -->""" + \
      out_legend + \
    """
    </g>

    <!-- pop-up notes -->""" + \
    out_graph + \
    """
  </g>
</svg>"""

    with open('output/' + file_name, 'w', encoding="utf-8") as fh:
        fh.write(svg_code)
              
    print(f"Done. Number of processed countries: {len(df)}")
```
