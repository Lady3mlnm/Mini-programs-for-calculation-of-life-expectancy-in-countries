```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
pd.options.display.max_columns = 50
pd.options.display.max_rows = 100
```

```python
dd_to_rus = {'USA': 'США',
             'China': 'Китай',
             'Russia': 'Россия',
             'Ukraine': 'Украина'}
```

```python
# load info from Rosstat
df = pd.read_csv('data/World_Bank_Group.csv', index_col=0)

print(df.shape)
df
```

```python
df.columns = df.columns.map(int)
df = df.transpose()
df
```

<br />
<br />

```python
# Comparison of Ukraine and Russia
plt.rcParams['figure.figsize'] = (20, 6)

year_min = df.index[0]
year_max = df.index[-1]

plt.title('Life expectancy in Ukraine and Russia', fontsize=14)

for row_name, color_line in zip(['Ukraine', 'Russia'],
                                ['darkorange', 'green']):  #darksalmon
    plt.plot(df.index, df[row_name], linestyle='-', color=color_line, linewidth=3, label=row_name)

plt.axvline(x=1991, color='lightgrey', linewidth=2, zorder=0)
plt.axvline(x=2014, color='lightgrey', linewidth=1, zorder=0)

plt.xlim(year_min, year_max)
plt.grid(color='wheat', linewidth=0.25)

labels_x = range(year_min, year_max+1)
labels_y = range(64, 75)
plt.xticks(labels_x, fontsize=14, rotation='vertical')
plt.yticks(labels_y, fontsize=12)
plt.legend(loc='upper left', fontsize=12)
plt.show()
```

```python
# Russian version of the graph
plt.title('Ожидаемая продолжительность жизни в Украине и России', fontsize=14)

for row_name, color_line in zip(['Ukraine', 'Russia'],
                                ['darkorange', 'green']):  #darksalmon
    plt.plot(df.index, df[row_name], linestyle='-', color=color_line, linewidth=3, label=dd_to_rus[row_name])

plt.axvline(x=1991, color='lightgrey', linewidth=2, zorder=0)
plt.axvline(x=2014, color='lightgrey', linewidth=1, zorder=0)

plt.xlim(year_min, year_max)
plt.grid(color='wheat', linewidth=0.25)

labels_x = range(year_min, year_max+1)
labels_y = range(64, 75)
plt.xticks(labels_x, fontsize=14, rotation='vertical')
plt.yticks(labels_y, fontsize=12)
plt.legend(loc='upper left', fontsize=12)
plt.show()
```
