```python
# load stats about longevity per year
df = pd.read_excel('data/Life expectancy in Turkey -2023.xls', skiprows=4, skipfooter=4) \
       .dropna(axis='columns', how='all')

df
```

```python
df.columns = ['',
              '2013_total', '2013_male', '2013_female',
              '2013-14_total', '2013-14_male', '2013-14_female',
              '2015-17_total', '2015-17_male', '2015-17_female',
              '2018-20_total', '2018-20_male', '2018-20_female',
              '2021-23_total', '2021-23_male', '2021-23_female'    
             ]

df = df.set_index('') \
       .rename(index={'Türkiye-Türkiye': 'Türkiye'})

df              
```

```python
df = df.loc[:, ['2015-17_total', '2015-17_male', '2015-17_female',
                '2018-20_total', '2018-20_male', '2018-20_female',
                '2021-23_total', '2021-23_male', '2021-23_female']]

df
```

```python
# I don't use here rounding since values have high accuracy
df.insert(loc=3, column='2015-17_fΔm', value=df['2015-17_female']-df['2015-17_male'])
df.insert(loc=4, column='Δ_years_1', value=df['2018-20_total']-df['2015-17_total'])

df.insert(loc=8, column='2018-20_fΔm', value=df['2018-20_female']-df['2018-20_male'])
df.insert(loc=9, column='Δ_years_2', value=df['2021-23_total']-df['2018-20_total'])

df.insert(loc=13, column='2021-23_fΔm', value=df['2021-23_female']-df['2021-23_male'])
df.insert(loc=14, column='Δ_years_sum', value=df['2021-23_total']-df['2015-17_total'])

df
```

```python
pd.options.display.max_rows = 100
pd.options.display.precision = 3

# def sort_df(df, by=['2021-23_total', '2021-23_male'], ascending=False):
def sort_df(df, by:list, ascending=False):
    df.sort_values(by=by, ascending=ascending, inplace=True)
    indexes = df.index.to_list()
    ls_change_order = ['Türkiye']    # 'İstanbul'
    indexes = list(el for el in indexes if el not in ls_change_order)
    indexes = ls_change_order + indexes
    return df.reindex(indexes)

df = sort_df(df, by=['2021-23_total', '2021-23_male'])
df
```

```python
# create code for placing info in Wikipedia
def create_table(df, file_header, lang='ru'):

    def if_value(x, prec=2):
        return '—' if math.isnan(x) else \
               f"{x:0.{prec}f}"  if x>=0 else \
               f"−{-x:0.{prec}f}"                #"{x:0.{prec}f}".format(x, prec)
    
    def chval(x, prec=2, *, add_par=''):  # change_value
        return 'style="padding-right:1em;"| —' if math.isnan(x) else \
               f'style="padding-right:1em;color:darkgreen;{add_par}"| {x:0.{prec}f}' if x>=0.005 else \
               f'style="padding-right:1em;color:crimson;{add_par}"| −{-x:0.{prec}f}' if x<=-0.005 else \
               f'style="padding-right:1em;color:darkgray;{add_par}"| {abs(x):0.{prec}f}'  # normal sorting is required '−' before 0.0 - bug in wiki

    def chval_bold(x, prec=2, *, add_par=''):  # change_value
        return '—' if math.isnan(x) else \
               f'style="padding-right:1em;color:darkgreen;{add_par}"| \'\'\'{x:0.{prec}f}\'\'\'' if x>0 else \
               f'style="padding-right:1em;color:crimson;{add_par}"| \'\'\'−{-x:0.{prec}f}\'\'\'' if x<0 else \
               f'style="padding-right:1em;color:darkgray;{add_par}"| \'\'\'{x:0.{prec}f}\'\'\''                 #  background:#fffae0;
    
    with open('design/' + file_header, mode='r', encoding="utf-8") as fh:
        table_header = fh.read()
    
    st = ''
    for i in range(len(df)):
        ser = df.iloc[i]
        if ser.name == 'Türkiye':
             st += '\n' + '|-class=static-row-header\n' + \
                  f'| \'\'\'{dd_replacement[ser.name][lang][0]}\'\'\' ' + \
                    f'||style="background:#e0ffd8;"| \'\'\'{if_value(ser["2015-17_total"], 2)}\'\'\' ' + \
                  f'||style="background:#eaf3ff;"| \'\'\'{if_value(ser["2015-17_male"], 2)}\'\'\' ' + \
                  f'||style="background:#fee7f6;"| \'\'\'{if_value(ser["2015-17_female"], 2)}\'\'\' ' + \
                  f'||style="background:#fff8dc;"| \'\'\'{if_value(ser["2015-17_fΔm"], 2)}\'\'\' ' + \
                  f'||{chval_bold(ser["Δ_years_1"], 2, add_par="border-left-width:2px;")} ' + \
                  f'||style="background:#e0ffd8;border-left-width:2px;"| \'\'\'{if_value(ser["2018-20_total"], 2)}\'\'\' ' + \
                  f'||style="background:#eaf3ff;"| \'\'\'{if_value(ser["2018-20_male"], 2)}\'\'\' ' + \
                  f'||style="background:#fee7f6;"| \'\'\'{if_value(ser["2018-20_female"], 2)}\'\'\' ' + \
                  f'||style="background:#fff8dc;"| \'\'\'{if_value(ser["2018-20_fΔm"], 2)}\'\'\' ' + \
                  f'||{chval_bold(ser["Δ_years_2"], 2, add_par="border-left-width:2px;")} ' + \
                  f'||style="background:#e0ffd8;border-left-width:2px;"| \'\'\'{if_value(ser["2021-23_total"], 2)}\'\'\' ' + \
                  f'||style="background:#eaf3ff;"| \'\'\'{if_value(ser["2021-23_male"], 2)}\'\'\' ' + \
                  f'||style="background:#fee7f6;"| \'\'\'{if_value(ser["2021-23_female"], 2)}\'\'\' ' + \
                  f'||style="background:#fff8dc;"| \'\'\'{if_value(ser["2021-23_fΔm"], 2)}\'\'\' ' + \
                  f'||{chval_bold(ser["Δ_years_sum"], 2, add_par="border-left-width:2px;")}'
        else:        
            st += '\n' + '|-\n' + \
                  f'| [[{dd_replacement[ser.name][lang][1]}|{dd_replacement[ser.name][lang][0]}]] ' + \
                  f'||style="background:#e0ffd8;"| \'\'\'{if_value(ser["2015-17_total"], 2)}\'\'\' ' + \
                  f'||style="background:#eaf3ff;"| {if_value(ser["2015-17_male"], 2)} ' + \
                  f'||style="background:#fee7f6;"| {if_value(ser["2015-17_female"], 2)} ' + \
                  f'||style="background:#fff8dc;"| {if_value(ser["2015-17_fΔm"], 2)} ' + \
                  f'||{chval(ser["Δ_years_1"], 2, add_par="border-left-width:2px;")} ' + \
                  f'||style="background:#e0ffd8;border-left-width:2px;"| \'\'\'{if_value(ser["2018-20_total"], 2)}\'\'\' ' + \
                  f'||style="background:#eaf3ff;"| {if_value(ser["2018-20_male"], 2)} ' + \
                  f'||style="background:#fee7f6;"| {if_value(ser["2018-20_female"], 2)} ' + \
                  f'||style="background:#fff8dc;"| {if_value(ser["2018-20_fΔm"], 2)} ' + \
                  f'||{chval(ser["Δ_years_2"], 2, add_par="border-left-width:2px;")} ' + \
                  f'||style="background:#e0ffd8;border-left-width:2px;"| \'\'\'{if_value(ser["2021-23_total"], 2)}\'\'\' ' + \
                  f'||style="background:#eaf3ff;"| {if_value(ser["2021-23_male"], 2)} ' + \
                  f'||style="background:#fee7f6;"| {if_value(ser["2021-23_female"], 2)} ' + \
                  f'||style="background:#fff8dc;"| {if_value(ser["2021-23_fΔm"], 2)} ' + \
                  f'||{chval(ser["Δ_years_sum"], 2, add_par="border-left-width:2px;")}'
            
    if lang == 'ru':
        st = re.sub('(?<=\\d)\\.(?=\\d)', ',', st)  # replace . to comma, if this . is between two digits

    st = table_header + st + '\n' + '|}'
    
    # gray color for missing values
    st = st.replace(';"| —', ';color:silver;"| —')

    return st


table_code = create_table(df, file_header='turkstat_header_en-2023.txt', lang='en')
with open('output/Table code for Turkish regions -en.txt', 'w', encoding="utf-8") as fh:
    fh.write(table_code)
    
table_code = create_table(df, file_header='turkstat_header_ru-2023.txt', lang='ru')
with open('output/Table code for Turkish regions -ru.txt', 'w', encoding="utf-8") as fh:
    fh.write(table_code)
```
