```python
def pretify_df(df):
    ls_remained = [
        'Russian Federation',
        'Estonia',
        'Latvia',
        'Lithuania',
        'Belarus',
        'Ukraine',
        'Moldova',
        'Georgia',
        'Azerbaijan',
        'Armenia',
        'Kazakhstan',
        'Uzbekistan',
        'Turkmenistan',
        'Kyrgyz Republic',
        'Tajikistan',
        'Turkiye',
        'United States',
        'Canada',
        'Korea, Rep.',
        "Korea, Dem. People's Rep.",
        'Spain',
        'Portugal'
    ]
    
    return df.loc[ls_remained] \
             .rename(index = {'Russian Federation': 'Russia',
                              'Kyrgyz Republic': 'Kyrgyzstan',
                              'Turkiye': 'Turkey',
                              'United States':'USA',
                              'Korea, Rep.': 'South Korea',
                              "Korea, Dem. People's Rep.": 'North Korea'
                             }) \
             .iloc[:, 3:-2] \
             .rename(columns=int)  # transform names of columns from strings to integers, like '1960' -> 1960
```

```python
ChartParams = namedtuple('ChartParams', ['ylim', 'empty_labels_down', 'empty_labels_up', 'legend_loc', 'ncol', 'label_x_rotation'],
                                        defaults=(None, 0, 0, 'upper left', 2, 'vertical'))

AnnotationNote = namedtuple('AnnotationNote', ['date', 'height', 'text_en', 'text_ru', 'fs', 'color', 'alpha'],  # fs - font_size
                            defaults=(2000, 80, 'name', 'имя', 15, 'dimgray', None))
```

```python
def rounding_up(x):
    rem = x % 1
    return (x // 1 + 0.25) if rem < 0.149 else \
           (x // 1 + 0.5) if rem < 0.399 else \
           (x // 1 + 1.25) if rem > 0.901 else \
           math.ceil(x)

def rounding_down(x):
    rem = x % 1
    return (x // 1 + 0.75) if rem > 0.851 else \
           (x // 1 + 0.5) if rem > 0.601 else \
           (x // 1 - 0.25) if rem < 0.099 else \
           math.floor(x)
```

```python
# Function that transform data in standard form (yyyy-mm-dd) in float number (yyyy.xx...)
def data_to_ration(data_st):
    year, month, day = (int(v) for v in data_st.split('-'))
    if year % 4 == 0:
        nmb_days_in_month = [31, 29, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
        no_day = sum(nmb_days_in_month[:month-1]) + day
        # print(f"no_day: {no_day}")
        ratio = year + (no_day - 1) / 366
    else:
        nmb_days_in_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
        no_day = sum(nmb_days_in_month[:month-1]) + day
        # print(f"no_day: {no_day}")
        ratio = year + (no_day - 1) / 365
    return ratio
```

```python
# Function that place a text fragment on chart
def place_text(plt, date_st, height_age, s, fs=18, color='dimgray', alpha=1):
    plt.text(x=data_to_ration(date_st)-0.3, y=height_age-0.25, s=s, verticalalignment='top', fontsize=fs, color=color, alpha=alpha)
```

```python
def create_chart(df_total_all, df_male_all, df_female_all, dd_regions, *, figure_size=(24, 15),
                 title_en='', title_ru='', USSR_naming=False, ver_lines=[], ver_lines_strong=[], chart_params=ChartParams(),
                 lang=LANG, destination=DESTINATION_OUTPUT, file_name=None):
    
    if destination == 'pass':
        return None
    
    plt.rcParams['figure.figsize'] = figure_size
    
    df_total = df_total_all.loc[dd_regions.keys()]
    df_male = df_male_all.loc[dd_regions.keys()]
    df_female = df_female_all.loc[dd_regions.keys()]
    
    year_min = df_total.index[0]
    year_max = df_total.index[-1]
   
    min_chart = min(df_total.min().min(), df_male.min().min(), df_female.min().min())
    max_chart = max(df_total.max().max(), df_male.max().max(), df_female.max().max())
    print(f"min: {min_chart:.2f}, max: {max_chart:.2f}      Δ: {(max_chart - min_chart):.2f}")
    
    # determine limits on axis
    if not chart_params.ylim:
        chart_params = chart_params._replace(ylim=(rounding_down(min_chart),
                                                   rounding_up(max_chart)
                                                  ) )
    
    ylim_bottom_label = math.ceil(chart_params.ylim[0])
    ylim_up_label = math.floor(chart_params.ylim[1])
    
    year_min = df_total.columns[0]
    year_max = df_total.columns[-1]

    # set chart areas, their locations, sizes and appearances
    fig = plt.figure()
    ax = fig.add_subplot(111)
    
    # set where ticks and labels will be shown around charts
    ax.tick_params(axis='x', which='both', top=False, bottom=True, labeltop=False, labelbottom=True, pad=1.5)
    ax.tick_params(axis='y', which='both', left=True, right=True, labelleft=True, labelright=True, pad=1.5)

    # set limits of chart axes
    ax.set_xlim(year_min, year_max)
    ax.set_ylim(chart_params.ylim[0], chart_params.ylim[1])

    
    # set ticks and labels
    labels_x = range(year_min, year_max+1)
    ax.set_xticks(labels_x)
    ax.set_xticklabels(labels_x, fontsize=14, rotation=chart_params.label_x_rotation)
    
    labels_y = tuple(range(ylim_bottom_label, ylim_up_label + 1))
    ax.set_yticks(labels_y)
    ax.set_yticklabels(('',) * chart_params.empty_labels_down + 
                        labels_y[chart_params.empty_labels_down:-chart_params.empty_labels_up if chart_params.empty_labels_up else None] +
                        ('',) * chart_params.empty_labels_up, fontsize=12)

    # Make grid. Since lines for years are shifted at this type of charts, not-standard way of grid creation is used
    for year in range(year_min, year_max+1):
        plt.axvline(x=year, color='wheat', linewidth=0.25 , zorder=0)
    for year in range(year_min - year_min % 10 + 10, year_max, 10):
        plt.axvline(x=year, color='wheat', linewidth=0.5 if year == 2000 else 0.333 , zorder=0)
    for age in range(ylim_bottom_label, ylim_up_label+1):
        plt.axhline(y=age, color='wheat', linewidth=0.25, zorder=0)
    for age in range(ylim_bottom_label - ylim_bottom_label % 5 + 5, ylim_up_label, 5):
        plt.axhline(y=age, color='wheat', linewidth=0.333, zorder=0)
     
    # place vertical lines and annotations for leaders
    for data in ver_lines:
        plt.axvline(x=data, color='lightgrey', linewidth=1.25, zorder=0)
    
    for data in ver_lines_strong:
        plt.axvline(x=data, color='lightgrey', linewidth=2.5, zorder=0)

  
    # set title
    ax.set_title(label=title_ru if lang=='ru' else title_en, fontsize=15)
    
    #———  setting for chart 1  ———
    for region, color in zip(dd_regions.keys(), dd_regions.values()):
        region_short = dd_to_rus[region] if lang=='ru' else region
        if USSR_naming:
            region_short = dd_to_USSR_naming[region]
        ax.plot(df_female.loc[region, :].index, df_female.loc[region, :].values, color=color, linestyle=':', linewidth=1.5, label=f"{region_short} -{'женщины' if lang=='ru' else 'female'}", zorder=12)
        ax.plot(df_total.loc[region, :].index, df_total.loc[region, :].values, color=color, linestyle='-', linewidth=3, label=f"{region_short} -{'общая' if lang=='ru' else 'overall'}", zorder=11)
        ax.plot(df_male.loc[region, :].index, df_male.loc[region, :].values, color=color, linestyle='--', linewidth=1.5, label=f"{region_short} -{'мужчины' if lang=='ru' else 'male'}", zorder=13)
     
    ax.legend(loc=chart_params.legend_loc, fontsize=12, ncol=chart_params.ncol)
        # print(region)
        # print(df_total.loc[region, :])
        # print()
    
    # show created chart or save it to file
    if destination == 'file':
        plt.savefig(f"group_with_genders_output/{file_name}{' -ru' if lang=='ru' else ''}.png",
                    bbox_inches='tight', facecolor='white', pad_inches=0.05)       
        print('Data has written to file')
    else:
        plt.show()


# LANG = 'en'
dd_regions = {'Ukraine': 'darkorange',
              'Russia': 'green'}    
ver_lines_strong = [1991]
ver_lines = [2014, 2019]
chart_params=ChartParams(ylim=(57.5, 78.5))

create_chart(
        df_total, df_male, df_female,
        dd_regions=dd_regions, figure_size=(30, 14),
        title_en='Life expectancy at birth in Ukraine and Russia',
        title_ru='Ожидаемая при рождении продолжительность жизни в Украине и России',
        ver_lines_strong=ver_lines_strong, ver_lines=ver_lines, chart_params=chart_params,
        lang=LANG, destination=DESTINATION_OUTPUT, file_name='Comparison of life expectancy in Ukraine and Russia')
```

```python
def create_chart_leader(df_total_all, df_male_all, df_female_all, dd_regions, *, figure_size=(24, 15),
                        title_en='', title_ru='', USSR_naming=False, ver_lines=[], ver_lines_strong=[], annotations=[], chart_params=ChartParams(),
                        lang=LANG, destination=DESTINATION_OUTPUT, file_name=None):
    
    if destination == 'pass':
        return None
    
    plt.rcParams['figure.figsize'] = figure_size

    df_total = df_total_all.loc[dd_regions.keys()]
    df_male = df_male_all.loc[dd_regions.keys()]
    df_female = df_female_all.loc[dd_regions.keys()]
    
    year_min = df_total.index[0]
    year_max = df_total.index[-1]
    
    min_chart = min(df_total.min().min(), df_male.min().min(), df_female.min().min())
    max_chart = max(df_total.max().max(), df_male.max().max(), df_female.max().max())
    print(f"min: {min_chart:.2f}, max: {max_chart:.2f}      Δ: {(max_chart - min_chart):.2f}")
    
    # determine limits on axis
    if not chart_params.ylim:
        chart_params = chart_params._replace(ylim=(rounding_down(min_chart),
                                                   rounding_up(max_chart)
                                                  ) )
    
    ylim_bottom_label = math.ceil(chart_params.ylim[0])
    ylim_up_label = math.floor(chart_params.ylim[1])
    
    year_min = df_total.columns[0]
    year_max = df_total.columns[-1]

    # set chart areas, their locations, sizes and appearances
    fig = plt.figure()
    ax = fig.add_subplot(111)
    
    # set where ticks and labels will be shown around charts
    ax.tick_params(axis='x', which='both', top=False, bottom=True, labeltop=False, labelbottom=True, pad=1.5)
    ax.tick_params(axis='y', which='both', left=True, right=True, labelleft=True, labelright=True, pad=1.5)

    # set limits of chart axes
    ax.set_xlim(year_min, year_max)
    ax.set_ylim(chart_params.ylim[0], chart_params.ylim[1])

    
    # set ticks and labels
    labels_x = range(year_min, year_max+1)
    ax.set_xticks(labels_x)
    ax.set_xticklabels(labels_x, fontsize=13.5, rotation=chart_params.label_x_rotation)
    
    labels_y = tuple(range(ylim_bottom_label, ylim_up_label + 1))
    ax.set_yticks(labels_y)
    ax.set_yticklabels(('',) * chart_params.empty_labels_down + 
                        labels_y[chart_params.empty_labels_down:-chart_params.empty_labels_up if chart_params.empty_labels_up else None] +
                        ('',) * chart_params.empty_labels_up, fontsize=11.5)

    # Make grid. Since lines for years are shifted at this type of charts, not-standard way of grid creation is used
    for year in range(year_min, year_max+1):
        plt.axvline(x=year - 0.5, color='wheat', linewidth=0.25 , zorder=0)
    for year in range(year_min - year_min % 10 + 10, year_max, 10):
        plt.axvline(x=year - 0.5, color='wheat', linewidth=0.5 if year == 2000 else 0.333 , zorder=0)
    for age in range(ylim_bottom_label, ylim_up_label+1):
        plt.axhline(y=age, color='wheat', linewidth=0.25, zorder=0)
    for age in range(ylim_bottom_label - ylim_bottom_label % 5 + 5, ylim_up_label, 5):
        plt.axhline(y=age, color='wheat', linewidth=0.333, zorder=0)
     
    # place vertical lines and annotations for leaders
    for data_st in ver_lines:
        data_ratio = data_to_ration(data_st)
        plt.axvline(x=data_ratio - 0.5, color='lightgrey', linewidth=1.25, zorder=0)
    
    for data_st in ver_lines_strong:
        data_ratio = data_to_ration(data_st)
        plt.axvline(x=data_ratio - 0.5, color='lightgrey', linewidth=2.5, zorder=0)

    for annotation in annotations:
        place_text(plt, annotation.date, annotation.height, annotation.text_ru if lang=='ru' else annotation.text_en, annotation.fs, annotation.color, annotation.alpha)

  
    # set title
    ax.set_title(label=title_ru if lang=='ru' else title_en, fontsize=15)
    
    #———  setting for chart 1  ———
    for region, color in zip(dd_regions.keys(), dd_regions.values()):
        region_short = dd_to_rus[region] if lang=='ru' else region
        if USSR_naming:
            region_short = dd_to_USSR_naming[region]
        ax.plot(df_female.loc[region, :].index, df_female.loc[region, :].values, color=color, linestyle=':', linewidth=1.5, label=f"{region_short}-{'женщины' if lang=='ru' else 'female'}", zorder=12)
        ax.plot(df_total.loc[region, :].index, df_total.loc[region, :].values, color=color, linestyle='-', linewidth=3, label=f"{region_short}-{'общая' if lang=='ru' else 'overall'}", zorder=11)
        ax.plot(df_male.loc[region, :].index, df_male.loc[region, :].values, color=color, linestyle='--', linewidth=1.5, label=f"{region_short}-{'мужчины' if lang=='ru' else 'male'}", zorder=13)
        
    ax.legend(loc=chart_params.legend_loc, fontsize=12, ncol=chart_params.ncol)
        # print(region)
        # print(df_total.loc[region, :])
        # print()
    
    # show created chart or save it to file
    if destination == 'file':
        plt.savefig(f"output-USSR/{file_name}{' -ru' if lang=='ru' else ''}.png",
                    bbox_inches='tight', facecolor='white', pad_inches=0.05)       
        print('Data has written to file')
    else:
        plt.show()


dd_regions = {'Ukraine': 'darkorange',
              'Russia': 'green'}    
ver_lines_strong = ['1991-12-25']
ver_lines = ['1964-10-14', '1982-11-12', '1984-02-13', '1985-03-11', '1991-12-25', '2000-01-01', '2008-05-07', '2012-05-07']
annotations = [
    AnnotationNote(date='1960-08-01', height=74.7 if LANG=='ru' else 74.65, text_en='Khrushchev', text_ru='Хрущёв'), #75.15
    AnnotationNote(date='1969-11-00' if LANG=='ru' else '1969-01-00', height=76.1, text_en='Brezhnev', text_ru='Брежнев'),
    AnnotationNote(date='1982-09-15' if LANG=='ru' else '1982-09-10', height=76.15, text_en='Andropov', text_ru='Андропов'),
    AnnotationNote(date='1983-12-15', height=75.75, text_en='Chernenko', text_ru='Черненко'),
    AnnotationNote(date='1986-10-00', height=76.1, text_en='Gorbachev', text_ru='Горбачёв'),
    AnnotationNote(date='1992-01-00', height=76.15, text_en='Kravchuk', text_ru='Кравчук', color='darkorange', alpha=0.5),
    AnnotationNote(date='1992-01-00', height=75.75, text_en='Yeltsin', text_ru='Ельцин', color='green', alpha=0.5),
]
chart_params=ChartParams(ylim=(61, 76), empty_labels_up=1)

create_chart_leader(
        df_total.loc[:, YEAR_START:YEAR_END], df_male.loc[:, YEAR_START:YEAR_END], df_female.loc[:, YEAR_START:YEAR_END],
        dd_regions=dd_regions, figure_size=(22, 13.5),
        title_en='Life expectancy at birth in Ukrainian SSR and Russian SFSR in comparison with leaders of the country',
        title_ru='Ожидаемая при рождении продолжительность жизни в Украинской ССР и Российской СФСР в сопосталении с главами государства',
        USSR_naming=True, ver_lines_strong=ver_lines_strong, ver_lines=ver_lines, annotations=annotations, chart_params=chart_params,
        lang=LANG, destination=DESTINATION_OUTPUT, file_name='Comparison of life expectancy in UkSSR and RSFSR vs leader (WBG)')
```
