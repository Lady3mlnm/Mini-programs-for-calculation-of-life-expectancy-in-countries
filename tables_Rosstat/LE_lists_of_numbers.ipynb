#### Auxiliary program for create a list of numbers from Rosstat Excel file that can be required for creation of graphs with 3-party services.

```python
import pandas as pd
```

```python
pd.options.display.max_columns = 50
```

```python
# load stats about longevity per year
df = pd.read_excel('data/Rosstat_by_year_and_gender.xls', skiprows=2)

print(df.shape)
df.head(6)
```

```python
# drop redundant columns
df.drop(['Unnamed: 0', 'Unnamed: 2'], axis='columns', inplace=True)

# convert gender column to index
df.set_index('Unnamed: 1', inplace=True)

# for brevityh of display remove name of index column.
df.index.name = ''

print(df.shape)
df
```

```python
df = df.rename(index={
    'Женщины': 'females',
    'Мужчины': 'males',
    'Оба пола': 'overall'
}).reindex(['females', 'overall', 'males', ])

df
```

```python
print('Years:')
print(*df.columns.to_list(), sep=',')
```

```python
print('Females:')
print(*df.loc['females'].to_list(), sep=',')
```

```python
print('Overall:')
print(*df.loc['overall'].to_list(), sep=',')
```

```python
print('Males:')
print(*df.loc['males'].to_list(), sep=',')
```

```python
# write data to disk
df = df.transpose()
df.index.name = 'year'
df
```

```python
df.to_csv('output/Rosstat.scs')
print('Done')
```
