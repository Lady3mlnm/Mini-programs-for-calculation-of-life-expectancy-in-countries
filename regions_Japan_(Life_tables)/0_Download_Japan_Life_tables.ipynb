This script downloaded  "Life tables" from the [Japanese Mortality Database](https://www.ipss.go.jp/p-toukei/JMD/index-en.asp) <i>(ipss.go.jp)</i>

```python
import requests
import re
from time import sleep
```

```python
for nmb in range(48):
    for server_file, suffix in zip(["bltper", "mltper", "fltper"], ['total', 'males', 'females']):
        url = f"https://www.ipss.go.jp/p-toukei/JMD/{nmb:02}/STATS/{server_file}_1x1.txt"
        file_path = f"data-Life_tables/JpLT_{nmb:02}-{suffix}.csv"
    
        response = requests.get(url)
    
        if response.status_code == 200:
            with open(file_path, 'w') as file:
                text = response.text
    
                # Replace gaps from 3 to 6 characters to tabs. And also change end of lines for correct writing to file.
                text_transformed = re.sub(' {2,11}', '\t', text) \
                                     .replace('\r\n\t', '\r\n') \
                                     .replace('\r\n', '\n')
                file.write(text_transformed)
                
                print(f"{nmb:02}. {server_file}_1x1.txt is downloaded")
        else:
            raise Exception(f"Response 200 during processing of '{nmb} - {server_file}'")
    
        sleep(1)
```
