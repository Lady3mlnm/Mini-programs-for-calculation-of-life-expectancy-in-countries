```python
import pandas as pd
import math
import re

import sys
sys.path.append("..")
import mal_moduls_private.mal_total as mal
```

```python
df_state = pd.read_csv('data/Life expectancy at birth by region and sex -Sweden.csv', sep='\t', skiprows=2, index_col='region')

df_countries = pd.read_csv('data/Life expectancy at birth by region and sex -countries.csv', sep='\t', skiprows=2,
                           encoding='iso8859_10', index_col='region')  # iso8859_10 - coding for Nordic languages, iso8859_2 - for Central and Eastern Europe

df = pd.concat([df_state, df_countries])

df.index = df.index.map(lambda st: st[3:])
df.index.name = ''

df_men = df.loc[df.sex=='men'].drop(columns='sex')
df_men = df_men.sort_values(by=['2019-2023', '2018-2022'], ascending=False)
df_men = pd.concat([df_men.loc[['Sweden']], df_men.drop('Sweden')])

df_women = df.loc[df.sex=='women'].drop(columns='sex')
df_women = df_women.sort_values(by=['2019-2023', '2018-2022'], ascending=False)
df_women = pd.concat([df_women.loc[['Sweden']], df_women.drop('Sweden')])

del df_state, df_countries, df

df_mean = (df_women + df_men) / 2
df_mean = df_mean.sort_values(by=['2019-2023', '2018-2022'], ascending=False)
df_mean = pd.concat([df_mean.loc[['Sweden']], df_mean.drop('Sweden')])

df_diff = df_women - df_men
df_diff = df_diff.sort_values(by=['2019-2023', '2018-2022'], ascending=False)
df_diff = pd.concat([df_diff.loc[['Sweden']], df_diff.drop('Sweden')])
```

```python
# create code for placing info in Wikipedia
def create_table(df_men=df_men, df_women=df_women, df_mean=df_mean, df_diff=df_diff, dd_replacement=dd_replacement, file_header='', lang='en'):

    def if_value(x, prec=2):
        return '—' if math.isnan(x) else \
               f"{x:0.{prec}f}"  if x>=0 else \
               f"−{-x:0.{prec}f}"                #"{x:0.{prec}f}".format(x, prec)
    
    def chval(x, prec=2, *, add_par=''):  # change_value
        return f'style="color:silver;{add_par}"| —' if math.isnan(x) else \
               f'style="color:darkgreen;{add_par}"| {x:0.{prec}f}' if x>0 else \
               f'style="color:crimson;{add_par}"| −{-x:0.{prec}f}' if x<0 else \
               f'style="color:darkgray;{add_par}"| {x:0.{prec}f}'
    
    def chval_bold(x, prec=2, *, add_par=''):  # change_value
        return f'style="color:silver;{add_par}"| \'\'\'—\'\'\'' if math.isnan(x) else \
               f'style="color:darkgreen;{add_par}"| \'\'\'{x:0.{prec}f}\'\'\'' if x>0 else \
               f'style="color:crimson;{add_par}"| \'\'\'−{-x:0.{prec}f}\'\'\'' if x<0 else \
               f'style="color:darkgray;{add_par}"| \'\'\'{x:0.{prec}f}\'\'\''

    with open('design/' + file_header, mode='r', encoding="utf-8") as fh:
        table_header = fh.read()

    st = ''
    for i in range(len(df_men)):
        ser_men = df_men.iloc[i]
        ser_women = df_women.loc[ser_men.name]
        ser_mean = df_mean.loc[ser_men.name]
        ser_diff = df_diff.loc[ser_men.name]
        if ser_men.name in ["Sweden"]:
            st += '\n' + '|-class=static-row-header\n' + \
                  f'|style="text-align:left;"| \'\'\'{dd_replacement[ser_men.name][lang][0]}\'\'\' ' + \
                  f'||style="background:#eaf3ff;"| \'\'\'{if_value(ser_men["2014-2018"])}\'\'\' ' + \
                  f'||style="background:#fee7f6;"| \'\'\'{if_value(ser_women["2014-2018"])}\'\'\' ' + \
                  f'||style="background:#fff8dc;"| \'\'\'{if_value(ser_diff["2014-2018"])}\'\'\' ' + \
                  f'|| \'\'\'{if_value(ser_mean["2014-2018"])}\'\'\' ' + \
                  f'|| {chval_bold(ser_mean["2019-2023"]-ser_mean["2014-2018"], add_par="border-left-width:2px;")} ' + \
                  f'||style="background:#eaf3ff;border-left-width:2px;"| \'\'\'{if_value(ser_men["2019-2023"])}\'\'\' ' + \
                  f'||style="background:#fee7f6;"| \'\'\'{if_value(ser_women["2019-2023"])}\'\'\' ' + \
                  f'||style="background:#fff8dc;"| \'\'\'{if_value(ser_diff["2019-2023"])}\'\'\' ' + \
                  f'|| \'\'\'{if_value(ser_mean["2019-2023"], prec=2)}\'\'\''
        else:
            name_link = dd_replacement[ser_men.name][lang][1]
            name_visible = dd_replacement[ser_men.name][lang][0]
            name_inserted = name_link if name_link == name_visible else f"{name_link}|{name_visible}"
            st += '\n' + '|-\n' + \
                  f'| [[{name_inserted}]] ' + \
                  f'||style="background:#eaf3ff;"| {if_value(ser_men["2014-2018"])} ' + \
                  f'||style="background:#fee7f6;"| {if_value(ser_women["2014-2018"])} ' + \
                  f'||style="background:#fff8dc;"| {if_value(ser_diff["2014-2018"])} ' + \
                  f'|| {if_value(ser_mean["2014-2018"])} ' + \
                  f'|| {chval(ser_mean["2019-2023"]-ser_mean["2014-2018"], add_par="border-left-width:2px;")} ' + \
                  f'||style="background:#eaf3ff;border-left-width:2px;"| {if_value(ser_men["2019-2023"])} ' + \
                  f'||style="background:#fee7f6;"| {if_value(ser_women["2019-2023"])} ' + \
                  f'||style="background:#fff8dc;"| {if_value(ser_diff["2019-2023"])} ' + \
                  f'|| {if_value(ser_mean["2019-2023"], prec=2)}'


    if lang == 'ru':
        st = re.sub('(?<=\d)\.(?=\d)', ',', st)  # replace . to comma, if this . is between two digits
        # st = st.replace('padding-right:1,5ex;', 'padding-right:1.5ex;')

    st = table_header + st + '\n|}'
    
    # gray color for missing values
    # st = st.replace(';"|—', ';color:silver;"|—')

    return st


table_code = create_table(file_header='Swedish_header_en -2023.txt', lang='en')

# write the code to file
with open('output/Table code for Swedish regions -en.txt', 'w', encoding="utf-8") as fh:
    fh.write(table_code)
```
