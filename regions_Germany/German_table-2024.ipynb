```python
import pandas as pd
import math
import re

import sys
sys.path.append("..")
import mal_moduls_private.mal_total as mal
```

```python
df_male_1 = pd.read_csv('data/2021-23_male.csv', sep='\t', index_col='Länder')
df_male_1.index.name = ''
df_male_1.loc[['Berlin', 'Germany']]
```

```python
df_male_2 = pd.read_csv('data/2022-24_male.csv', sep='\t', index_col='Länder')
df_male_2.index.name = ''
df_male_2.loc[['Berlin', 'Germany']]
```

```python
df_female_1 = pd.read_csv('data/2021-23_female.csv', sep='\t', index_col='Länder')
df_female_1.index.name = ''
df_female_1.loc[['Berlin', 'Germany']]
```

```python
df_female_2 = pd.read_csv('data/2022-24_female.csv', sep='\t', index_col='Länder')
df_female_2.index.name = ''
df_female_2.loc[['Berlin', 'Germany']]
```

```python
# create summary DataFrame with all required data
df = pd.concat([df_male_1.loc[:, '0'],
                df_female_1.loc[:, '0'],
                (df_female_1.loc[:, '0'] - df_male_1.loc[:, '0']).round(2),       # ! without rounding it is possible error during sorting
                ((df_male_1.loc[:, '0'] + df_female_1.loc[:, '0']) / 2).round(2),
                ((df_male_2.loc[:, '0'] + df_female_2.loc[:, '0'] - df_male_1.loc[:, '0'] - df_female_1.loc[:, '0']) / 2).round(2),
                df_male_2.loc[:, '0'],
                df_female_2.loc[:, '0'],
                (df_female_2.loc[:, '0'] - df_male_2.loc[:, '0']).round(2),       # ! without rounding it is possible error during sorting
                ((df_male_2.loc[:, '0'] + df_female_2.loc[:, '0']) / 2).round(2)
               ],
                  axis='columns',
                  keys=['male_1', 'female_1', 'fΔm_1', 'ar_mean_1', 'Δ_periods', 'male_2', 'female_2', 'fΔm_2', 'ar_mean_2'])


# sort DataFrame
df.sort_values(by=['ar_mean_2', 'male_2'], ascending=False, inplace=True)

# place record about country as a whole to top of DataFrame
indexes = df.index.to_list()
ls_change_order = ['Germany']
indexes = list(el for el in indexes if el not in ls_change_order)
indexes = ls_change_order + indexes
df = df.reindex(indexes)

print(len(df))

df
```

```python
# create code for placing info in Wikipedia
def create_table(df, file_header, lang='ru'):

    def if_value(x, prec=2):
        return '—' if math.isnan(x) else \
               f"{x:0.{prec}f}"  if x>=0 else \
               f"−{-x:0.{prec}f}"                #"{x:0.{prec}f}".format(x, prec)
    
    def chval(x, prec=2, *, add_par=''):  # change_value
        return f' —' if (math.isnan(x) & (add_par=='')) else \
               f'style="{add_par}"| —' if math.isnan(x) else \
               f'style="color:darkgreen;{add_par}"| {x:0.{prec}f}' if x>0 else \
               f'style="color:crimson;{add_par}"| −{-x:0.{prec}f}' if x<0 else \
               f'style="color:darkgray;{add_par}"| {x:0.{prec}f}'
    
    def chval_bold(x, prec=2, *, add_par=''):  # change_value
        return ' \'\'\'—\'\'\'' if (math.isnan(x) & (add_par=='')) else \
               f'style="{add_par}"| \'\'\'—\'\'\' ' if math.isnan(x) else \
               f'style="color:darkgreen;{add_par}"| \'\'\'{x:0.{prec}f}\'\'\'' if x>0 else \
               f'style="color:crimson;{add_par}"| \'\'\'−{-x:0.{prec}f}\'\'\'' if x<0 else \
               f'style="color:darkgray;{add_par}"| \'\'\'{x:0.{prec}f}\'\'\''
    
    with open('design/' + file_header, mode='r', encoding="utf-8") as fh:
        table_header = fh.read()

    st = ''
    for i in range(len(df)):
        ser = df.iloc[i]
        if ser.name in ["Germany"]:
            st += '\n' + '|-class=static-row-header\n' + \
                 f'| \'\'\'{dd_replacement[ser.name][lang][0]}\'\'\' ' + \
                 f'||style="background:#eaf3ff;"| \'\'\'{if_value(ser["male_1"])}\'\'\' ' + \
                 f'||style="background:#fee7f6;"| \'\'\'{if_value(ser["female_1"])}\'\'\' ' + \
                 f'||style="background:#fff8dc;"| \'\'\'{if_value(ser["fΔm_1"])}\'\'\' ' + \
                 f'|| \'\'\'{if_value(ser["ar_mean_1"], prec=2)}\'\'\' ' + \
                 f'||{chval_bold(ser["Δ_periods"], add_par="border-left-width:2px;")} ' + \
                 f'||style="background:#eaf3ff;border-left-width:2px;"| \'\'\'{if_value(ser["male_2"])}\'\'\' ' + \
                 f'||style="background:#fee7f6;"| \'\'\'{if_value(ser["female_2"])}\'\'\' ' + \
                 f'||style="background:#fff8dc;"| \'\'\'{if_value(ser["fΔm_2"])}\'\'\' ' + \
                 f'|| \'\'\'{if_value(ser["ar_mean_2"])}\'\'\''
        else:
            name_link = dd_replacement[ser.name][lang][1]
            name_visible = dd_replacement[ser.name][lang][0]
            name_inserted = name_link if name_link == name_visible else f"{name_link}|{name_visible}"
            st += '\n' + '|-\n' + \
                  f'| [[{name_inserted}]] ' + \
                  f'||style="background:#eaf3ff;"| {if_value(ser["male_1"])} ' + \
                  f'||style="background:#fee7f6;"| {if_value(ser["female_1"])} ' + \
                  f'||style="background:#fff8dc;"| {if_value(ser["fΔm_1"])} ' + \
                  f'|| {if_value(ser["ar_mean_1"])} ' + \
                  f'||{chval(ser["Δ_periods"], add_par="border-left-width:2px;")} ' + \
                  f'||style="background:#eaf3ff;border-left-width:2px;"| {if_value(ser["male_2"])} ' + \
                  f'||style="background:#fee7f6;"| {if_value(ser["female_2"])} ' + \
                  f'||style="background:#fff8dc;"| {if_value(ser["fΔm_2"])} ' + \
                  f'|| {if_value(ser["ar_mean_2"])}'

    if lang == 'ru':
        st = re.sub('(?<=\\d)\\.(?=\\d)', ',', st)  # replace . to comma, if this . is between two digits

    st = table_header + st + '\n|}'
    
    # gray color for missing values
    st = st.replace(';"|—', ';color:silver;"|—')

    return st


table_code = create_table(df, file_header='German_header_en.txt', lang='en')

# write the code to file
with open('output/Table code for German states -2024 -en.txt', 'w', encoding="utf-8") as fh:
    fh.write(table_code)
```
